```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pseudocolor plot
u = np.linspace(-2,2,65)
v = np.linspace(-1,1,33)
X,Y = np.meshgrid(u,v)
Z = X**2/25 + Y**2/4
plt.pcolor(Z)
plt.show()
```

```python
plt.pcolor(Z)
plt.colorbar()
plt.show()
```

```python
plt.pcolor(Z, cmap = 'spring')
plt.colorbar()
plt.show()
```

```python
plt.pcolor(Z, cmap = 'summer')
plt.colorbar()
plt.axis('tight')
plt.show()
```

```python
#X,Y are 2D meshgrid
plt.pcolor(X,Y,Z)
plt.colorbar()
plt.show()
```

```python
plt.contour(Z)
plt.show()
```

```python
plt.contour(Z, 40)
plt.show()
```

```python
plt.contourf(X,Y,Z,20)
plt.colorbar()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.contourf(Z, 20)
```

```python
plt.subplot(2,2,2)
plt.contourf(Z, 10)
```

```python
plt.subplot(2,2,1)
plt.contourf(Z, 10, cmap = 'spring')

plt.subplot(2,2,2)
plt.contourf(Z, 10, cmap = 'summer')

plt.subplot(2,2,3)
plt.contourf(Z, 10, cmap = 'autumn')

plt.subplot(2,2,4)
plt.contourf(Z, 10, cmap = 'winter')

plt.tight_layout()
plt.show()
```
