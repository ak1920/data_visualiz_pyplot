```python
import matplotlib.pyplot as plt

plt.figure()
# Set x-axis range
plt.xlim((1,9))
# Set y-axis range
plt.ylim((1,9))
# Draw lines to split quadrants
plt.plot([5,5],[1,9], linewidth=4, color='red' )
plt.plot([1,9],[5,5], linewidth=4, color='red' )
plt.title('Quadrant plot')
# Draw some sub-regions in upper left quadrant
plt.plot([3,3],[5,9], linewidth=2, color='blue')
plt.plot([1,5],[7,7], linewidth=2, color='blue')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('/Users/aslihankoksoy/Downloads/tips.csv')
```

```python
df.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, 
       hue ='sex', palette = 'Set1')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df,
          col = 'smoker')
plt.show()
```

```python
sns.residplot(x='total_bill', y='tip', data=df, color = 'magenta')
plt.show()
```

```python
# Generate a scatter plot of bill and tip using red circles
plt.scatter(df['total_bill'], df['tip'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between bill and tip
sns.regplot(x='total_bill', y='tip', data=df, scatter=None, color='blue', label='First Order')

# Plot in green a linear regression of order 2 between bill and tip
sns.regplot(x='total_bill', y='tip', data=df, scatter=None, order=2, color='green', label='Second Order')
```

```python
#lineer regr beyween bill and tip, row-wise by smoking
sns.lmplot(x='total_bill', y='tip', data=df, row='smoker')

# Display the plot
plt.show()
```

```python
sns.stripplot(y = 'tip', data = df)
plt.ylabel('tip$')
plt.show()
```

```python
sns.stripplot(x = 'time', y = 'tip', data = df)
plt.ylabel('tip$')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'tip', data = df)
plt.ylabel('tip$')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'tip', data = df, hue = 'sex')
plt.ylabel('tip$')
plt.show()
```

```python
sns.violinplot(x='day', y='tip', 
               data = df, inner=None, color = 'lightgray') 
sns.stripplot(x='day', y='tip', data=df, size = 4, jitter =True)
plt.ylabel('tip$')
plt.show()
```
