```python
# making line plots

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,1000)
```

```python
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('myplot')
plt.show()
```

```python
# making scatter plot
#first is for x axis(horiz), second is for y axis(vert)
x = np.random.randn(100,2)
plt.scatter(x[:,0], x[:,1]) 
            
```

```python
x = np.random.randn(200,2)
x[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(x[:,0], x[:,1], c = y) 
```
